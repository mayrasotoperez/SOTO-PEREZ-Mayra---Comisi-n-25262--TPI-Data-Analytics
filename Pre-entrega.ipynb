### 5- Calidad de Datos: Identificar valores nulos y duplicados en los conjuntos de datos. Documentar el estado inicial de los datos.

```python
def ident_nul_dup(df, nombredf, columna_clave=None):

  print("-----------------------------------")
  print("########## ", nombredf, " ############") #muestra el nombre
  print("-----------------------------------\n")
  print("Valores Nulos por columna:\n----------------------")
  print(df.isnull().sum()) #para contar valores nulos.

  print('\nCant. filas duplicadas:', df.duplicated(keep=False).sum()) #para contar filas duplicados (keep=False para marcar como True todas las filas que tienen otra igual)
  print("")

  if columna_clave and columna_clave in df.columns :    # ingresa solo si se pasa una columna y dicha columna est√° en dicho df.

        # Contar cu√°ntas filas tienen valores repetidos en esa columna
        dup_key = df[columna_clave].duplicated(keep=False).sum()
        print(f"Duplicados por clave '{columna_clave}':", dup_key, "\n")

        # Si existen duplicados, mostrar cu√°les son los valores m√°s repetidos
        if dup_key > 0:
            duplicados_ordenados = (
                df[df[columna_clave].duplicated(keep=False)][columna_clave] #filtrar el df para que te devuelva solo los duplicados
                .value_counts()# Cuenta cu√°ntas veces aparece cada valor
                .sort_values(ascending=False)   # Ordena de mayor a menor (m√°s duplicados arriba)
            )

            print("\nüîÅ Top 5 valores duplicados m√°s frecuentes:")
            # Mostrar solo los primeros 5 (los m√°s repetidos)
            display(duplicados_ordenados.head(5))
        else:
            print(f"No se encontraron duplicados en la clave '{columna_clave}'.\n")
  else:
        # Si la clave no fue pasada o no existe en el DataFrame
        if columna_clave:
            print(f"La clave '{columna_clave}' no existe en el DataFrame.\n")
        else:
            print("No se indic√≥ una clave para analizar duplicados por columna.\n")


```

```python
ident_nul_dup(clientes,"CLIENTES RAW","id_cliente")
ident_nul_dup(ventas,"VENTAS RAW", "id_venta")
ident_nul_dup(marketing,"MARKETING RAW", "id_campanha")
```

## Etapa 2: Preprocesamiento y Limpieza de Datos (Clases 5 a 8)

### 1- Limpieza de Datos: Limpiar el conjunto de datos eliminando duplicados y caracteres no deseados. Documentar el proceso y los resultados.

```python
# Crear copias independientes para no modificar los originales

ventas_clean = ventas.copy()
clientes_clean = clientes.copy()
marketing_clean = marketing.copy()
```

```python
# Eliminar filas completamente duplicadas

clientes_clean = clientes_clean.drop_duplicates()
ventas_clean = ventas_clean.drop_duplicates()
marketing_clean = marketing_clean.drop_duplicates()

# Chequear el estado
ident_nul_dup(clientes_clean,"CLIENTES","id_cliente")
ident_nul_dup(ventas_clean,"VENTAS", "id_venta")
ident_nul_dup(marketing_clean,"MARKETING", "id_campanha")
```

```python
# Funci√≥n para limpiar texto en columnas tipo string

def normalizar_texto(df):

    for col in df.select_dtypes(include="object").columns:
        # Se agrupan las operaciones entre par√©ntesis () para escribirlas en varias l√≠neas
        # Python eval√∫a todo el bloque como una √∫nica expresi√≥n.
        df[col] = (
            df[col]
            .astype(str) # .astype(str)  ‚Üí convierte todo a texto
            .str.strip() # .str.strip()  ‚Üí Elimina espacios al inicio y final
            .str.replace(r"[\u200b\t\r\n]", "", regex=True)
            # .str.replace(patron, reemplazo, regex=True)
            #   patron: expresi√≥n regular que busca caracteres invisibles (\u200b, tabulaciones, saltos)
            #   reemplazo: ""  ‚Üí los elimina
            #   regex=True indica que 'patron' es una expresi√≥n regular.
            .str.replace(" +", " ", regex=True)
            # reemplaza "uno o m√°s espacios consecutivos" por un solo espacio
            .str.title() # Convierte a T√≠tulo: "juan p√©rez" ‚Üí "Juan P√©rez"
        )
        return df

clientes_clean = normalizar_texto(clientes_clean)
ventas_clean = normalizar_texto(ventas_clean)
marketing_clean = normalizar_texto(marketing_clean)
```

```python
# Funci√≥n para Normalizar fechas

# Si alguna columna contiene fechas (por ejemplo "fecha" o "fechanotificacion"), se intenta convertir a formato datetime de pandas.
# to_datetime intenta interpretar el formato y transforma valores inv√°lidos en NaT (Not a Time).

def norm_fechas(df):
  contador = 0
  for col in df.columns:
    if "fecha" in col.lower():  # detecta columnas con la palabra "fecha"
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
            # Par√°metros:
            #   errors="coerce" ‚Üí convierte valores no v√°lidos en NaT (evita error)
            #   dayfirst=True   ‚Üí interpreta formatos tipo "DD/MM/YYYY" (formato latino)
            print("Se normaliz√≥ la columna:",col)
            contador += 1
  if contador == 0: print("No se encontraron columnas de fecha para normalizar")
  print("------------------------")
  return df



clientes_clean = norm_fechas(clientes_clean)
ventas_clean = norm_fechas(ventas_clean)
marketing_clean = norm_fechas(marketing_clean)
```

```python
# Normalizar valores num√©ricos

# Campo "precio"

def norm_precio_pesos (df):
  for col in df.columns:
    if "precio" in col.lower():  # detecta columnas con la palabra "precio"
      df[col] = df[col].astype(str).str.replace("$", "", regex=False).str.replace(",", "", regex=False).str.strip()
              # .astype(str) Convierte todo a texto
              # .str.replace("$", "", regex=False) Elimina el s√≠mbolo $
              #  regex=False ‚Üí interpreta "$" literalmente, no como expresi√≥n regular
              # .str.replace(",", "", regex=False) Elimina comas de miles 1,000  1000
              # .str.strip()  Quita espacios sobrantes

      df[col] = pd.to_numeric(df[col], errors="coerce")
      # pd.to_numeric convierte texto a n√∫mero (float o int)
      # Par√°metros: errors="coerce" ‚Üí reemplaza valores no convertibles con NaN
  return df

#  Campo "cantidad"

def norm_cantidad(df):
  for col in df.columns:
    if "cantidad" in col.lower():
        df[col] = pd  .to_numeric(df[col], errors="coerce")  .astype("Int64")
        # .astype("Int64") usa el tipo entero de pandas que permite valores nulos (NaN)
  return df

clientes_clean = norm_cantidad(norm_precio_pesos(clientes_clean))
ventas_clean = norm_cantidad(norm_precio_pesos(ventas_clean))
marketing_clean = norm_cantidad(norm_precio_pesos(marketing_clean))
```

Reporte Global luego de la limpieza de datos


```python
# ============================================
# üìä REPORTE GLOBAL DE CALIDAD DE DATOS
# ============================================
# Esta funci√≥n lee los tres DataFrames limpios (o los recibe en memoria)
# y muestra un resumen comparativo de nulos, duplicados y tipos de datos.
# ============================================

def reporte_calidad_global(dfs, nombres):
    """
    Crea un resumen de calidad de varios DataFrames.

    Par√°metros:
      dfs: lista de DataFrames (por ejemplo [ventas_clean, clientes_clean, marketing_clean])
      nombres: lista de nombres correspondientes (["VENTAS", "CLIENTES", "MARKETING"])
    """
    resumen = []
    #zip-->es una funci√≥n incorporada de Python que une elementos de dos (o m√°s) iterables
    # ‚Äîpor ejemplo, listas, tuplas o cualquier objeto iterable‚Äî en pares ordenados.
    for df, nombre in zip(dfs, nombres):
        nulos = df.isna().sum().sum()                    # Total de valores nulos, no por columnas sino total, por eso el doble sum
        duplicados = df.duplicated(keep='last').sum()    # Total de filas duplicadas | 'last' -> ya que quiero que coincidan los duplicados por las filas eliminadas por duplicados
        columnas = len(df.columns)                       # Cantidad de columnas
        filas = len(df)                                  # Cantidad de registros

        resumen.append({
            "Dataset": nombre,
            "Filas": filas,
            "Columnas": columnas,
            "Nulos totales": nulos,
            "Duplicados": duplicados,
        })

    reporte = pd.DataFrame(resumen)
    #display(reporte)
    return reporte


# ============================================

print(reporte_calidad_global([ventas, clientes, marketing], ["Ventas Raw", "Clientes Raw", "Marketing Raw"]))

print(reporte_calidad_global([ventas_clean, clientes_clean, marketing_clean], ["Ventas Clean", "Clientes Clean", "Marketing Clean"]))
```
